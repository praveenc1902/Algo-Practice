```python

def handle_pay_cols(df2):
    for i in range(1,7):
        fil =  eval(f"df2.PAY_{i} <0")
        df2.loc[fil,f"PAY_{i}"] = 0
    return df2

df3 = handle_pay_cols(df2)

```

```python
df3.PAY_2.value_counts()
```

```python
df3.head()
```

```python
cat_vars = ['SEX','EDUCATION','MARRIAGE','PAY_1','PAY_2','PAY_3','PAY_4',
                   'PAY_5','PAY_6']
df2[cat_vars] = df2[cat_vars].astype(str)
df3= pd.get_dummies(df2,cat_vars)
```

```python
df3.info()
```

```python
df3.columns = df3.columns.map(lambda x:x.lower())
```

```python
df3.columns
```

```python
col_to_norm = ['limit_bal', 'age', 'bill_amt1', 'bill_amt2', 'bill_amt3', 'bill_amt4',
       'bill_amt5', 'bill_amt6', 'pay_amt1', 'pay_amt2', 'pay_amt3',
       'pay_amt4', 'pay_amt5', 'pay_amt6']
df3[col_to_norm] = df3[col_to_norm].apply(lambda x: x-np.mean(x)/np.std(x))
```

```python
df3.head()
```

```python
x = df3.drop('def_pay',axis = 1)
y = df3.def_pay
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
